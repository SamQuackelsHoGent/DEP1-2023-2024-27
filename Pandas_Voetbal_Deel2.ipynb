# Cleaning up klassement.csv file

```python
# Import pandas
import pandas as pd
```

### Creating a dataframe

```python
# creeÃ«r een dataframe

# kolom namen defineren
columns = ["seizoen","speeldag","stand","ploeg","aantal_gespeelde_matchen","aantal_gewonnen_matchen","aantal_gelijkgespeelde_matchen",
           "aantal_verloren_matchen","doelpunten","puntenverschil","punten"]

# methode read_csv()
# params:
#   - naam van de csv file
#   - header=None -> Als je dit niet doet valt de eerste row weg
df = pd.read_csv("klassementen.csv", header=None)

# voeg de kolom namen toe
df.columns = columns
```

```python
# Toon de eerste paar rijen
df.head()
```

```python
# Toon wat generieke informatie
df.info()
```

### Removing the text from the day column

```python
# Verwijder onodige tekst
for dag in range(1, 35):
  df.loc[df["speeldag"] == f"{dag}.speeldag", "speeldag"] = dag

df.loc[df["speeldag"]=="Wedstrijden van de 1e speeldag Eerste klasse 60/61", "speeldag"] = "31"
df.loc[df["speeldag"]=="Datum", "speeldag"] = "32"
df.loc[df["speeldag"]=="Tijd", "speeldag"] = "33"
df.loc[df["speeldag"]=="Thuisploeg", "speeldag"] = "34"
```

```python
# Toon de unieke values van de speeldag kolom
df.speeldag.unique()
```

```python
# Verander de dtype van object naar int
df.speeldag = df.speeldag.astype('int64')
df.speeldag.dtype
```

### Change the season to the right type

```python
# Maak twee kolommen voor de gemaakte goals en de geincaseerde goals
df[["year1", "year2"]] = df.seizoen.str.split("/", expand=True)
```

```python
df.year1 = df.year1.astype('int64')
```

```python
df["fixed_seizoen"] = df["year1"]+1900
```

```python
df["fixed_seizoen"][df["fixed_seizoen"]<1960] = df["fixed_seizoen"].loc[df["fixed_seizoen"]<1960]+100
```

```python
df.fixed_seizoen.unique()
```

```python
df.head()
```

### Recalculate all the pointscores using the newer 3 point system

```python
# maak een nieuwe kolom voor het driepunten systeem
df["driePuntenSysteem"] = (df.aantal_gewonnen_matchen)*3 + (df.aantal_gelijkgespeelde_matchen)*1
```

```python
# Toon de unieke values van de driepunten systeem kolom
df.driePuntenSysteem.unique()
```

```python
# Toon de eerste paar rijen
df.head()
```

### Recalculate with the 2 point system

```python
# maak een nieuwe kolom voor de positieve punten van het twee punten systeem
df["tweePuntenSysteemWins"] = (df.aantal_gewonnen_matchen)*2 + (df.aantal_gelijkgespeelde_matchen)*1
```

```python
# Toon de unieke values van de positieve twee punte kolom
df.tweePuntenSysteemWins.unique()
```

```python
# maak een nieuwe kolom voor de negatieve punten van het twee punten systeem
df["tweePuntenSysteemLosses"] = (df.aantal_verloren_matchen)*2 + (df.aantal_gelijkgespeelde_matchen)*1
```

```python
# Toon de unieke values van de negatieve twee punte kolom
df.tweePuntenSysteemLosses.unique()
```

```python
# Toon de eerste paar rijen
df.head()
```

### Adding the made goals

```python
# Maak twee kolommen voor de gemaakte goals en de geincaseerde goals
df[["madeGoals", "recievedGoals"]] = df.doelpunten.str.split(":", expand=True)
```

```python
# Verander de dtypes naar int
df.madeGoals = df.madeGoals.astype('int64')
df.recievedGoals = df.recievedGoals.astype('int64')
```

```python
# Toon de eerste paar rijen
df.head()
```

### Adding the team id

```python
# Lees het stamnummer bestand in
columns = ["club_id", "ploeg", "stamnummer", "roepnaam"]
id_df = pd.read_csv("stamnummers.csv", sep=";", header=None)
id_df.columns = columns
```

```python
# Drop duplicate ploegen
id_df.drop(index=115, inplace=True)
id_df.drop(index=24, inplace=True)
id_df.drop(index=48, inplace=True)
id_df.drop(index=131, inplace=True)
```

```python
# Merge de twee dataframes on de ploeg kolom
df = pd.merge(df, id_df, on="ploeg", how="left")
```

### Check ofdat een ploeg evenveel gespeeld heeft dan dat er speeldagen zijn

```python
# add een kolomn die checkt ofdat speeldag verschillend is van het aantal gespeelde matchen
df["speeldagCheck"] = df["speeldag"]==df["aantal_gespeelde_matchen"]
```

```python
# Checkt de unique value counts
df.speeldagCheck.value_counts()

# Seizoen 75/76 -> Niet elke ploeg speelt op Ã©Ã©n speeldag
# Seizoen 02/03 -> Niet elke ploeg speelt op Ã©Ã©n speeldag
# Seizoen 09/10 -> Moeskroen heeft 0 matchen gespeeld doorheen het seizoen
```

### Check ofdat het doelpuntensaldo klopt

```python
# Maak een nieuwe boolean kolom die checkt ofdat het doelpunten saldo klopt
df["puntenVerschilCheck"] = df["puntenverschil"]==df["madeGoals"]-df["recievedGoals"]
```

```python
# check ofdat er false values zijn -> Nope
df.puntenVerschilCheck.value_counts()
```

```python
# Toon de eerste paar rijen
df.head()
```

### Verwijder de onodige kolommen

```python
# Toon de kolommen
df.info()
```

```python
# Drop onodigde kolommen
df = df.drop(columns="doelpunten")
df = df.drop(columns="punten")
df = df.drop(columns="club_id")
df = df.drop(columns="ploeg")
df = df.drop(columns="speeldagCheck")
df = df.drop(columns="puntenVerschilCheck")
df = df.drop(columns="year1")
df = df.drop(columns="year2")
df = df.drop(columns="seizoen")
```

```python
# Toon de resterende kolommen
df.info()
```

### Schrijft de dataframe naar een nieuw csv file

```python
# Herangschik alle kolommen
desired_order = ["fixed_seizoen", "speeldag", "stand", "stamnummer", "roepnaam", "aantal_gespeelde_matchen", "aantal_gewonnen_matchen",
                 "aantal_gelijkgespeelde_matchen", "aantal_verloren_matchen", "madeGoals", "recievedGoals", "puntenverschil",
                 "tweePuntenSysteemWins", "tweePuntenSysteemLosses", "driePuntenSysteem"]

df = df[desired_order]
```

```python
# Toon de nieuwe orde
df.head()
```

```python
# transform de dataframe in een csv file

# methode to_csv
# Params:
#   - index=False -> Geen extra index kolom nodig
#   - header=None -> Geen kolom namen nodig

df.to_csv('klassementen_fixed.csv', index=False, header=None)
```
